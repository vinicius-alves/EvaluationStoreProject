```python
df_agrupado = df.groupby(['Estado','DscDetalheMercado','DatCompetencia'], as_index = False )['VlrMercado'].sum()
```

```python
df_agrupado['DatCompetencia']  = pd.to_datetime(df_agrupado['DatCompetencia'])
```

```python
df_pivot = df_agrupado.pivot_table(index = ['Estado','DatCompetencia'], values = 'VlrMercado',columns = 'DscDetalheMercado', fill_value = 0).reset_index()
```

```python
df_export = df_pivot.drop(columns = ['energia consumida','impostos']).reset_index(drop = True)
df_export = df_export.rename(columns = {'energia consumida TE':'EnergiaConsumida',
                                        'DatCompetencia':'MesReferencia',
                                        'energia injetada':'EnergiaInjetada',
                                        'número de consumidores':'QtdConsumidores',
                                        'receita':'Receita'})

df_export.to_csv('dados_consumo_agrupados_nivel_estado.csv', sep = ';', index = False, decimal = ',')
```

```python
estados_sudeste = ['Rio de Janeiro','São Paulo','Minas Gerais']

for coluna in [ 'energia consumida TE', 'energia injetada']:
    fig = px.area(df_pivot[df_pivot['Estado'].isin(estados_sudeste)], x = 'DatCompetencia', y = coluna, color = 'Estado', title = coluna)
    fig.show()
```

```python
df_agrup = df_pivot.groupby('DatCompetencia', as_index = False).agg({'receita':'sum', 'número de consumidores':'sum','energia consumida TE':'sum'})

df_agrup['ticket médio'] = df_agrup['receita']/df_agrup['número de consumidores']
df_agrup['take rate'] =  df_agrup['energia consumida TE']/ df_agrup['receita']
df_agrup['consumo médio'] =  df_agrup['energia consumida TE']/ df_agrup['número de consumidores']

for coluna in ['ticket médio','take rate','consumo médio']:
    fig = px.line(df_agrup, x = 'DatCompetencia', y = coluna,  title = coluna)
    fig.show()
```

```python
with open('gis-dataset-brasil\\uf\\geojson\\uf.json', encoding='latin1') as f:
    geojson_raw = f.readlines()

geojson = json.loads(''.join(geojson_raw))
df_pivot['NOME_UF'] = df_pivot['Estado']
```

```python
df_pivot['NOME_UF'] = df_pivot['Estado']
filtro_ultimo_mes = df_pivot['DatCompetencia'] == df_pivot['DatCompetencia'].max()

df_pivot['ticket médio'] = df_pivot['receita']/df_pivot['número de consumidores']
df_pivot['take rate'] =  df_pivot['energia consumida TE']/ df_pivot['receita']
df_pivot['consumo médio'] =  df_pivot['energia consumida TE']/ df_pivot['número de consumidores']

for coluna in ['ticket médio','take rate','consumo médio','número de consumidores']:

    fig = px.choropleth_mapbox(df_pivot[filtro_ultimo_mes], geojson=geojson, locations='NOME_UF', color=coluna, title = coluna,
                            color_continuous_scale="Viridis",
                            mapbox_style="carto-positron",
                            featureidkey = 'properties.NOME_UF',
                            zoom=3, 
                            opacity=0.5
                            )
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()
```
